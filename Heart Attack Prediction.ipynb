```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output = False , drop = 'first')
encoder = ohe.fit_transform(df[['Sex']])
encoded_sex = pd.DataFrame(encoder, columns = ohe.get_feature_names_out(['Sex']))
df = pd.concat([df,encoded_sex] , axis = 1)
```

```python
df.head(5)
```

```python
attack_counts = df.groupby(['HadHeartAttack', 'State']).size().reset_index(name='Count')
sns.barplot(x='HadHeartAttack', y='Count', hue='State', data=attack_counts)
```

```python
x = df.drop(columns = ['State' , 'HeightInMeters', 'WeightInKilograms','HadSkinCancer',
                       'HadDepressiveDisorder','HadKidneyDisease','HadDiabetes','BlindOrVisionDifficulty',
                        'ECigaretteUsage','RaceEthnicityCategory','HIVTesting','TetanusLast10Tdap','Sex'
                      ,'PatientID','GeneralHealth','AgeCategory','BMI','HadCOPD','SmokerStatus','PneumoVaxEver',
                      'DeafOrHardOfHearing' ,'DifficultyConcentrating', 'HighRiskLastYear','CovidPos','FluVaxLast12'  
                      ]
           
           )
y = df['HadStroke']
```

```python
x
```

```python
x.dtypes
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
```

```python
from imblearn.under_sampling import RandomUnderSampler
under_sample = RandomUnderSampler(sampling_strategy = 'auto')
x_sample,y_sample = under_sample.fit_resample(x,y)
```

```python
x_sample = x_sample[:min(len(x_sample) , len(y_sample))]
x_sample.shape
```

```python
y_sample = y_sample[:min(len(x_sample) , len(y_sample))]
y_sample.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_sample,y_sample,random_state = 42,test_size = 0.2)
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
logreg_model = LogisticRegression(max_iter=100)
xgb_model = XGBClassifier()

```

```python
x_test.shape
```

```python
y_test.shape
```

```python
rf_model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = rf_model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
logreg_model.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = logreg_model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
xgb_model.fit(x_train,y_train)
```

```python
y_pred = xgb_model.predict(x_test)
accuracy = accuracy_score(y_test,y_pred)
accuracy
```
